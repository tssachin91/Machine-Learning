# Convolutional Neural Network

### Importing the libraries

```
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```
tf.__version__
```

## Part 1 - Data Preprocessing

### Preprocessing the Training set

```
# Will perform all the transformations on the training set
# rescale => feature scaling 255 pixels between 0-1
train_datagen = ImageDataGenerator(
        rescale = 1./255,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True)

# Access the Data Directory 'dataset' Folder-> 'training_set'
# Training Size
# class_mode='binary' / 'categorical'
training_set = train_datagen.flow_from_directory(
        'dataset/training_set',
        target_size = (64, 64),
        batch_size = 32,
        class_mode='binary')
```

### Preprocessing the Test set

```
# Image transformation not performed to ensure no loss in data
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory(
        'dataset/test_set',
        target_size = (64, 64),
        batch_size = 32,
        class_mode='binary')
```

## Part 2 - Building the CNN

### Initialising the CNN

```
cnn = tf.keras.models.Sequential()
```

### Step 1 - Convolution

```
# Adding a Convolution Layer
# filters => No. of Feature Detectors
# kernel_size => Dimensions of the Feature Detector
# Activation => 'relu' => Rectifier Activation
# input_shape => 64*64 as we have transformed the training data & if B&W => Last Parameter = 1
cnn.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, activation = 'relu', input_shape = [64, 64, 3]))
```

### Step 2 - Pooling

```
# pool_size => Dimensions of Pool Feature Map
# strides => Movement in Feature Map when mapping Pool Feature Data
# padding => ignore empty column and fill dummy data
cnn.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
```

### Adding a second convolutional layer

```
cnn.add(tf.keras.layers.Conv2D(filters = 32, kernel_size = 3, activation = 'relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size = 2, strides = 2))
```

### Step 3 - Flattening

```
cnn.add(tf.keras.layers.Flatten())
```

### Step 4 - Full Connection

```
# units => No. of hidden neurons 
cnn.add(tf.keras.layers.Dense(units = 128, activation = 'relu'))
```

### Step 5 - Output Layer

```
cnn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

## Part 3 - Training the CNN

### Compiling the CNN

```
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### Training the CNN on the Training set and evaluating it on the Test set

```
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```

## Part 4 - Making a single prediction

```
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('dataset/single_prediction/cat_or_dog_1.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)

# data needs to provided in batches & for that purpose axis = 0
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)

# Identifying the formatting for Dog = 1, Cat = 0
training_set.class_indices
if result[0][0] == 1:
  prediction ='Dog'
else
  prediction ='Cat'
```
